```
# Uploading the Telecom Dataset2
# Telecom dataset2 will save it as df2.
#
df2= pd.read_csv('/content/Telcom_dataset2.csv')
print(df2)
df2.head(5)
```

```
# we raname some columns to be consistent.
#
df2.rename(columns ={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'},inplace=True)
df2.head(5)
```

2. Here, we are merging our datasets

```
# merging the df1, df2 
# use the function merge()
#
df12 = pd.merge(df1, df2, how ='outer')
df12
```

```
#Lets see descriptive statistics for merged data df12
#
df12.describe()
```

```
# adding one set of data to another
# will be done by calling the concat() function.
#
df1_2 = pd.concat([df1,df2], sort = False)
df1_2
```

```
#Lets see descriptive statistics for of concat data
#
df1_2.describe()
```

```
# lets merge the last dataset df3 with df12.
#
df123 = pd.merge(df12,df3, how = 'outer')
df123
```

```
#Lets see descriptive statistics for of merged data df123
#
df123.describe()
```

```
# adding one set of data to another
# will be done by calling the concat() function.
# let concat the dataset
# df
df1_2_3 = pd.concat([df1_2,df3], sort = False)
df1_2_3
```

3. Data cleaning

```
# we firt identify the missing values in the dataset
# by use of function null()
#
df123. isnull()
df123.fillna(df1_2_3.mean(), inplace=True)
df123.head(5)
```
